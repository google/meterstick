```python
import pandas as pd
df = pd.read_csv("https://raw.githubusercontent.com/google/meterstick/refs/heads/master/publications/a_grammar_of_data_analysis/datasets/baseball_pitches_to_SEA_batters.csv")
df
```

```python
player_names = pd.read_csv("https://raw.githubusercontent.com/google/meterstick/refs/heads/master/publications/a_grammar_of_data_analysis/datasets/baseball_player_names.csv")
player_names["Name"] = player_names["first_name"] + " " + player_names["last_name"]
df = df.merge(player_names[["id", "Name"]], how="left",
              left_on="pitcher_id", right_on="id")
df = df.merge(player_names[["id", "Name"]], how="left",
              left_on="batter_id", right_on="id", suffixes=("_pitcher", "_batter"))
df
```

```python
from meterstick import *

pitch_types = Count("pitch_type") | Distribution(over="pitch_type")
pitch_types.compute_on(df, split_by=["Name_batter", "Name_pitcher"])
```
